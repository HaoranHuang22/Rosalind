# 1. Counting DNA Nucleotides

```python
# dictionaries are useful for storing the count.
s = input()
dict_n = {}
for i in s:
    if i not in dict_n.keys():
        dict_n[i] = 1
    else:
        dict_n[i] += 1
print(dict_n['A'], dict_n['C'], dict_n['G'], dict_n['T'])
```

# 2. Transcrbing DNA into RNA


```python
# using replace function
t = input()
print(t.replace("T","U"))
```

# 3. Complementing a Strand of DNA

```python
# Firstly, reverse the input string(slice [::-1]), then, change the nucleotides with A-T, G-C rules.
s = input()
sc = s[::-1]
list_sc = []

for i in sc:
    if i == "A":
        list_sc.append("T")
    elif i == "T":
        list_sc.append("A")
    elif i == "C":
        list_sc.append("G")
    elif i == "G":
        list_sc.append("C")
        
sc = "".join(list_sc) # list to str
print(sc)
```
